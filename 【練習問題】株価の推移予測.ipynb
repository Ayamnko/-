```python
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
import seaborn as sns

# データの読み込み
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# flgを追加して結合
train_df['flg'] = 0
test_df['flg'] = 1
combined_df = pd.concat([train_df, test_df])



### 特徴量の生成 ###



# 元のデータフレームの保持
flg_column = combined_df['flg']
date_column = combined_df['Date']
up_column = combined_df['Up']

# 基本の列のリスト
columns = ['Close', 'Open', 'High', 'Low']

# 四則演算の組み合わせ
arithmetics_df = pd.DataFrame()
for i in range(len(columns)):
    for j in range(len(columns)):
        if i != j:
            col1 = columns[i]
            col2 = columns[j]


            # 加算
            arithmetics_df[f'{col1}_plus_{col2}'] = combined_df[col1] + combined_df[col2]

            # 減算
            arithmetics_df[f'{col1}_minus_{col2}'] = combined_df[col1] - combined_df[col2]

            # 乗算
            arithmetics_df[f'{col1}_times_{col2}'] = combined_df[col1] * combined_df[col2]

# 移動平均を計算する列のリスト
columns_to_average = combined_df.drop(columns=['Date', 'Up', 'flg'], axis=1).columns

# 各列に対して10日間の移動平均を計算
rolling_means_df = pd.DataFrame()
for col in columns_to_average:
    rolling_means_df[f'{col}_rolling_mean_10'] = combined_df[col].rolling(window=10).mean()

# 遅延特徴量
lag_days = [1, 5, 10]  # 遅延の時間幅
lags_df = pd.DataFrame()
for col in columns:
    for lag in lag_days:
        lags_df[f'{col}_lag_{lag}'] = combined_df[col].shift(lag)

# 差分特徴量
diffs_df = pd.DataFrame()
for col in columns:
    diffs_df[f'{col}_diff'] = combined_df[col].diff()
    diffs_df[f'{col}_diff_rolling_mean_10'] = diffs_df[f'{col}_diff'].rolling(window=10).mean()

# 統計的特徴量
stats_df = pd.DataFrame()
for col in columns:
    stats_df[f'{col}_max_10'] = combined_df[col].rolling(window=10).max()
    stats_df[f'{col}_min_10'] = combined_df[col].rolling(window=10).min()
    stats_df[f'{col}_median_10'] = combined_df[col].rolling(window=10).median()

# 指数移動平均（EMA）
span = 10  # EMAのスパン
ema_df = pd.DataFrame()
for col in columns:
    ema_df[f'{col}_ema_{span}'] = combined_df[col].ewm(span=span, adjust=False).mean()

# ボラティリティ指標
volatility_df = pd.DataFrame()
for col in columns:
    volatility_df[f'{col}_std_10'] = combined_df[col].rolling(window=10).std()

# トレンド指標
trend_df = pd.DataFrame()
for col in columns:
    trend_df[f'{col}_trend'] = (combined_df[col] > combined_df[col].shift(1)).astype(int)
    trend_df[f'{col}_trend_count'] = trend_df.groupby((trend_df[f'{col}_trend'] != trend_df[f'{col}_trend'].shift()).cumsum()).cumcount() + 1

# データの比率
ratios_df = pd.DataFrame()
ratios = [('Close', 'Open'), ('High', 'Low')]
for col1, col2 in ratios:
    ratios_df[f'{col1}_to_{col2}'] = combined_df[col1] / combined_df[col2].replace(0, pd.NA)

# Date列をdatetime型に変換
combined_df['Date'] = pd.to_datetime(combined_df['Date'])

# 年、月、日、曜日、四半期、週番号、日数を抽出
combined_df['year'] = combined_df['Date'].dt.year
combined_df['month'] = combined_df['Date'].dt.month
combined_df['day'] = combined_df['Date'].dt.day
combined_df['day_of_week'] = combined_df['Date'].dt.dayofweek
combined_df['quarter'] = combined_df['Date'].dt.quarter
combined_df['week_of_year'] = combined_df['Date'].dt.isocalendar().week
combined_df['day_of_year'] = combined_df['Date'].dt.dayofyear

# 各列に対して月ごとの移動平均を計算
monthly_means_df = pd.DataFrame()
for col in columns_to_average:
    monthly_means_df[f'{col}_monthly_rolling_mean'] = combined_df.groupby('month')[col].transform(lambda x: x.rolling(window=30, min_periods=1).mean())

# 各列に対して日ごとの移動平均を計算
daily_means_df = pd.DataFrame()
for col in columns_to_average:
    daily_means_df[f'{col}_daily_rolling_mean'] = combined_df.groupby('day')[col].transform(lambda x: x.rolling(window=7, min_periods=1).mean())

# 各列に対して月ごとの変動率を計算
monthly_changes_df = pd.DataFrame()
for col in columns_to_average:
    monthly_changes_df[f'{col}_monthly_change'] = combined_df.groupby('month')[col].transform(lambda x: x.pct_change())

# 各列に対して日ごとの変動率を計算
daily_changes_df = pd.DataFrame()
for col in columns_to_average:
    daily_changes_df[f'{col}_daily_change'] = combined_df.groupby('day')[col].transform(lambda x: x.pct_change())

# 多項式特徴量（2次と3次）
polynomial_df = pd.DataFrame()
for col in columns:
    polynomial_df[f'{col}_squared'] = combined_df[col] ** 2
    polynomial_df[f'{col}_cubed'] = combined_df[col] ** 3

# 累積統計量
cumulative_stats_df = pd.DataFrame()
for col in columns:
    cumulative_stats_df[f'{col}_cumsum'] = combined_df[col].cumsum()
    cumulative_stats_df[f'{col}_cummax'] = combined_df[col].cummax()
    cumulative_stats_df[f'{col}_cummin'] = combined_df[col].cummin()

# ヒストグラムビン
bins_df = pd.DataFrame()
for col in columns:
    bins_df[f'{col}_bin'] = pd.cut(combined_df[col], bins=10, labels=False)

# ウィンドウベースの統計量
window_size = 10
window_stats_df = pd.DataFrame()
for col in columns:
    window_stats_df[f'{col}_rolling_std_{window_size}'] = combined_df[col].rolling(window=window_size).std()

# 周期的な特徴量
periodic_df = pd.DataFrame()
for col in columns:
    periodic_df[f'{col}_quarter_sin'] = np.sin(2 * np.pi * combined_df['quarter'] / 4)
    periodic_df[f'{col}_quarter_cos'] = np.cos(2 * np.pi * combined_df['quarter'] / 4)
    periodic_df[f'{col}_month_sin'] = np.sin(2 * np.pi * combined_df['month'] / 12)
    periodic_df[f'{col}_month_cos'] = np.cos(2 * np.pi * combined_df['month'] / 12)
    periodic_df[f'{col}_day_sin'] = np.sin(2 * np.pi * combined_df['day_of_year'] / 365)
    periodic_df[f'{col}_day_cos'] = np.cos(2 * np.pi * combined_df['day_of_year'] / 365)

# 移動中央値
rolling_median_df = pd.DataFrame()
for col in columns:
    rolling_median_df[f'{col}_rolling_median_10'] = combined_df[col].rolling(window=10).median()

# 累積変化量
cumulative_changes_df = pd.DataFrame()
for col in columns:
    cumulative_changes_df[f'{col}_cumulative_change'] = combined_df[col].cumsum().diff()

# 上昇・下降の速度
speed_df = pd.DataFrame()
for col in columns:
    speed_df[f'{col}_up_speed'] = combined_df[col].diff().clip(lower=0)
    speed_df[f'{col}_down_speed'] = -combined_df[col].diff().clip(upper=0)

# 特徴量間の相関
correlation_df = pd.DataFrame()
correlation_matrix = combined_df[columns].corr()
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        col1 = columns[i]
        col2 = columns[j]
        correlation_df[f'{col1}_vs_{col2}_correlation'] = correlation_matrix.loc[col1, col2]

rolling_means_10_df = pd.DataFrame()
rolling_means_20_df = pd.DataFrame()
rolling_means_30_df = pd.DataFrame()
rolling_means_50_df = pd.DataFrame()
for col in columns_to_average:
    rolling_means_10_df[f'{col}_rolling_mean_10'] = combined_df[col].rolling(window=10).mean()
    rolling_means_20_df[f'{col}_rolling_mean_20'] = combined_df[col].rolling(window=20).mean()
    rolling_means_30_df[f'{col}_rolling_mean_30'] = combined_df[col].rolling(window=30).mean()
    rolling_means_50_df[f'{col}_rolling_mean_50'] = combined_df[col].rolling(window=50).mean()

# EMAのバリエーション（20日、30日、50日）
ema_20_df = pd.DataFrame()
ema_30_df = pd.DataFrame()
ema_50_df = pd.DataFrame()
for col in columns:
    ema_20_df[f'{col}_ema_20'] = combined_df[col].ewm(span=20, adjust=False).mean()
    ema_30_df[f'{col}_ema_30'] = combined_df[col].ewm(span=30, adjust=False).mean()
    ema_50_df[f'{col}_ema_50'] = combined_df[col].ewm(span=50, adjust=False).mean()

# ラグ間の差分
lag_diff_df = pd.DataFrame()
for col in columns:
    lag_diff_df[f'{col}_lag_1_vs_lag_5_diff'] = lags_df[f'{col}_lag_1'] - lags_df[f'{col}_lag_5']
    lag_diff_df[f'{col}_lag_5_vs_lag_10_diff'] = lags_df[f'{col}_lag_5'] - lags_df[f'{col}_lag_10']

# 年間の周期性
periodic_year_df = pd.DataFrame()
for col in columns:
    periodic_year_df[f'{col}_year_sin'] = np.sin(2 * np.pi * combined_df['year'] / 10)
    periodic_year_df[f'{col}_year_cos'] = np.cos(2 * np.pi * combined_df['year'] / 10)

# 四分位数統計量
quantiles_df = pd.DataFrame()
for col in columns:
    quantiles_df[f'{col}_25_quantile'] = combined_df[col].rolling(window=10).quantile(0.25)
    quantiles_df[f'{col}_75_quantile'] = combined_df[col].rolling(window=10).quantile(0.75)

# 短期および長期の移動平均
rolling_means_5_df = pd.DataFrame()
rolling_means_10_df = pd.DataFrame()
rolling_means_15_df = pd.DataFrame()
rolling_means_40_df = pd.DataFrame()
rolling_means_60_df = pd.DataFrame()
for col in columns_to_average:
    rolling_means_5_df[f'{col}_rolling_mean_5'] = combined_df[col].rolling(window=5).mean()
    rolling_means_10_df[f'{col}_rolling_mean_10'] = combined_df[col].rolling(window=10).mean()
    rolling_means_15_df[f'{col}_rolling_mean_15'] = combined_df[col].rolling(window=15).mean()
    rolling_means_40_df[f'{col}_rolling_mean_40'] = combined_df[col].rolling(window=40).mean()
    rolling_means_60_df[f'{col}_rolling_mean_60'] = combined_df[col].rolling(window=60).mean()

# 非線形変換
log_transforms_df = pd.DataFrame()
sqrt_transforms_df = pd.DataFrame()
for col in columns:
    log_transforms_df[f'{col}_log'] = np.log1p(combined_df[col])
    sqrt_transforms_df[f'{col}_sqrt'] = np.sqrt(combined_df[col].clip(lower=0))

# 高次統計量
high_order_stats_df = pd.DataFrame()
for col in columns:
    high_order_stats_df[f'{col}_std_10'] = combined_df[col].rolling(window=10).std()
    high_order_stats_df[f'{col}_skew'] = combined_df[col].rolling(window=10).apply(lambda x: pd.Series(x).skew())
    high_order_stats_df[f'{col}_kurtosis'] = combined_df[col].rolling(window=10).apply(lambda x: pd.Series(x).kurtosis())

# 特徴量間の追加の相互作用項
interaction_df_2 = pd.DataFrame()
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        col1 = columns[i]
        col2 = columns[j]
        interaction_df_2[f'{col1}_times_{col2}_squared_interaction'] = (combined_df[col1] * combined_df[col2]) ** 2
        interaction_df_2[f'{col1}_divided_by_{col2}_squared_interaction'] = (combined_df[col1] / combined_df[col2].replace(0, pd.NA)) ** 2

# 季節性のサイン・コサイン変換
seasonal_periodic_df = pd.DataFrame()
for col in columns:
    seasonal_periodic_df[f'{col}_month_sin'] = np.sin(2 * np.pi * combined_df['month'] / 12)
    seasonal_periodic_df[f'{col}_month_cos'] = np.cos(2 * np.pi * combined_df['month'] / 12)
    seasonal_periodic_df[f'{col}_week_sin'] = np.sin(2 * np.pi * combined_df['week_of_year'] / 52)
    seasonal_periodic_df[f'{col}_week_cos'] = np.cos(2 * np.pi * combined_df['week_of_year'] / 52)

# すべての特徴量を結合
combined_df = pd.concat([
    combined_df[['Date', 'flg', 'Up']],
    arithmetics_df,
    rolling_means_df,
    rolling_means_5_df,
    rolling_means_10_df,
    rolling_means_15_df,
    rolling_means_20_df,
    rolling_means_30_df,
    rolling_means_40_df,
    rolling_means_50_df,
    rolling_means_60_df,
    lags_df,
    diffs_df,
    stats_df,
    ema_df,
    ema_20_df,
    ema_30_df,
    ema_50_df,
    volatility_df,
    trend_df,
    ratios_df,
    monthly_means_df,
    daily_means_df,
    monthly_changes_df,
    daily_changes_df,
    polynomial_df,
    cumulative_stats_df,
    bins_df,
    window_stats_df,
    periodic_df,
    periodic_year_df,
    rolling_median_df,
    cumulative_changes_df,
    speed_df,
    correlation_df,
    quantiles_df,
    lag_diff_df,
    log_transforms_df,
    sqrt_transforms_df,
    high_order_stats_df,
    seasonal_periodic_df
], axis=1)




### データの選択 ###



# Dateとflg以外の列を選択
df_filtered = combined_df.drop(columns=['Date', 'flg'])

# 全ての列との相関係数を計算
correlation_matrix = df_filtered.corr()

# Upとの相関係数を取得
up_corr = correlation_matrix['Up']

# Upとの相関係数が0.1以上の列を取得（Upは除外）
filtered_columns = up_corr[abs(up_corr) >= 0.09].index
filtered_columns = filtered_columns.difference(['Up'])

# Date, flg, Upも含める
remaining_columns = filtered_columns.union(['Date', 'flg', 'Up'])

# combined_dfから対象の列を選択
combined_df = combined_df[remaining_columns]


### データの再分割 ###
train_df = combined_df[combined_df['flg'] == 0]
test_df = combined_df[combined_df['flg'] == 1]

# データの確認
display(train_df.head())
print(train_df.shape)
display(test_df.head())
print(test_df.shape)

# 基本統計量
display(train_df.describe())
display(test_df.describe())

# 欠損値の確認
print(train_df.isnull().sum())
print(test_df.isnull().sum())



# 相関行列を計算
correlation_matrix = train_df.drop(columns=['Date']).corr()

# ヒートマップを作成
plt.figure(figsize=(30, 28))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
'''
#「High」ごとに「Up」が0の行の数をカウント
count_up_zero = train_df[train_df['Up'] == 0].groupby('High').size().reset_index(name='Up_zero_count')

# 折れ線グラフを作成
plt.figure(figsize=(10, 6))
plt.plot(count_up_zero['High'], count_up_zero['Up_zero_count'], marker='o', linestyle='-')
plt.xlabel('High')
plt.ylabel('Count of Up = 0')
plt.title('Count of Up = 0 by High')
plt.grid(True)
plt.show()

# 各列に対して処理を行う
for column in train_df.columns:
    if column != 'Up':  # 'Up'列以外で処理を行う
        count_up_zero = train_df[train_df['Up'] == 0].groupby(column).size().reset_index(name='Up_zero_count')

        # 折れ線グラフを作成
        plt.figure(figsize=(10, 6))
        plt.plot(count_up_zero[column], count_up_zero['Up_zero_count'], marker='o', linestyle='-')
        plt.xlabel(column)
        plt.ylabel('Count of Up = 0')
        plt.title(f'Count of Up = 0 by {column}')
        plt.grid(True)
        plt.show()

# 結果を保存する辞書
stats_summary = {}

# 各列の値に対応した `Up` の統計量を計算
for column in train_df.columns:
    if column != 'Up':  # `Up` 列は除外
        # 各列のユニークな値ごとに `Up` の統計量を計算
        stats_per_value = train_df.groupby(column)['Up'].describe()

        # サンプル数 (count) を取得して追加
        stats_per_value['count'] = train_df.groupby(column)['Up'].count()

        stats_summary[column] = stats_per_value

# 結果の表示
for column, stats_df in stats_summary.items():
    print(f'\nStatistics of Up for each unique value in column "{column}":')
    display(stats_df)
'''
```

```python

# 各列0で欠損値を埋める
train_df.fillna(0, inplace=True)
test_df.fillna(0, inplace=True)

import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

# データの準備
X = train_df.drop(['Up', 'Date', 'flg'], axis=1)
y = train_df['Up']

# ロジスティック回帰モデルの定義
lr_model = LogisticRegression(
    random_state=42,
    solver='lbfgs',  # 小規模データセット向けのソルバー
    max_iter=1000
)

# クロスバリデーションの設定
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# RFEの評価結果を保存するリスト
feature_counts = []
accuracies = []

# 特徴量数の範囲（例として10から全特徴量数までの範囲を設定）
for num_features in range(X.shape[1] - 24, X.shape[1] + 1):
    # RFEによる特徴量選択
    selector = RFE(estimator=lr_model, n_features_to_select=num_features, step=1)
    selector = selector.fit(X, y)

    # 選択された特徴量
    X_selected = selector.transform(X)

    # クロスバリデーション
    cv_results = []
    for train_index, val_index in kf.split(X_selected, y):
        X_train, X_val = X_selected[train_index], X_selected[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]

        # ロジスティック回帰モデルの訓練
        lr_model.fit(X_train, y_train)

        # 検証データで予測
        val_predictions = lr_model.predict(X_val)

        # Accuracyの計算
        accuracy = accuracy_score(y_val, val_predictions)
        cv_results.append(accuracy)

    # クロスバリデーションの平均精度
    avg_accuracy = sum(cv_results) / len(cv_results)

    # 特徴量数と平均精度を記録
    feature_counts.append(num_features)
    accuracies.append(avg_accuracy)

    print(f'Features: {num_features}, Average Validation Accuracy: {avg_accuracy:.4f}')

# 結果をプロット
plt.figure(figsize=(10, 6))
plt.plot(feature_counts, accuracies, marker='o', linestyle='-', color='b')
plt.title('RFE Feature Selection vs. Model Accuracy')
plt.xlabel('Number of Features')
plt.ylabel('Validation Accuracy')
plt.grid(True)
plt.show()


```
